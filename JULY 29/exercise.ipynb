Student Scores Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_id = '1v4XQjb5sBWR8KDJ7qumf01u-CH3oBn17'
url = f'https://drive.google.com/uc?id={file_id}'
```

```python

df = pd.read_csv(url)
print(df.head())
```

1)Read it into a pandas DataFrame and print the first 3 rows.


```python
first_3_row=df.head(3)
print(first_3_row)
```

2. Calculate Average Scores
Add a new column Average that calculates the mean score for Math, Science, and
English for each student.



```python
df=df.assign(Average=(df['Math']+df['Science']+df['English'])/3)
print(df)
```

Print the student with the highest average.


```python

highest_avg = df[df['Average'] == df['Average'].max()]
print(highest_avg)
```

3. Filter Students
Display all students who have scored more than 80 in Science.


```python
df=df[df['Science']>80]
print(df)
```

Display all students with an average score below 70.

```python
df=df[df['Average']<70]
print(df)
```

4. Sorting
Sort the students by English score (descending) and print the top 3.


```python
df=df.sort_values(by='English',ascending=False).head(3)
print(df)
```

5. Chart Exercise
Create a bar chart showing studentsâ€™ names on the X-axis and their average
scores on the Y-axis.

```python
import matplotlib.pyplot as plt

plt.bar(df['Name'], df['Average'])
plt.xlabel('Name')
plt.ylabel('Average Score')
plt.title('Students Average Scores')
plt.show()
```

Create a line chart for all studentsâ€™ scores in Math, Science, and English (use
df.plot() ).


```python
import pandas as pd
import matplotlib.pyplot as plt
df.set_index('Name')[['Math', 'Science', 'English']].plot(kind='line', marker='o')
plt.title('Students Scores in Maths, Science , English')
plt.xlabel('Student Name')
plt.ylabel('Score')
plt.tight_layout()
plt.show()
```

Bonus Task
Find the subject with the highest total score across all students.



```python

totals = df[['Math', 'Science', 'English']].sum()

top_subject = totals.idxmax()
top_score = totals.max()

print(f"Subject with the highest total score: {top_subject} ({top_score} marks)")
```
