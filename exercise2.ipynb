**Exercise: Monthly Retail Sales Analysis**

1. Create a CSV of Sales Data

Save this as sales_data.csv in Colab.

Read it using pd.read_csv() and print the full DataFrame.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

file_id = '1DQxp9yzN_u4t9nNxzvpX2-C8XsUpYKdS'
url = f'https://drive.google.com/uc?id={file_id}'
```

```python

df = pd.read_csv(url)
print(df)
```

2. Total Sales Per Store

Group the data by Store and calculate total sales for each store.




```python
tot_sales_per_Store = df.groupby('Store').sum()
print(tot_sales_per_Store)
```

Identify which store had higher total sales.

```python
highest_totalsales=tot_sales_per_Store['Sales'].max()
print(highest_totalsales)
```

3. Monthly Sales Trend

Group the data by Month and sum up the total sales (Store A + Store B).

Print the result.

```python
tot_sal_by_month=df.groupby('Month').sum()
print(tot_sal_by_month)
```

4. Chart: Monthly Sales by Store

Create a line chart showing monthly sales for both Store A and Store B.

X-axis: Month

Y-axis: Sales

Two lines: Store A and Store B

```python

line_df = df.pivot(index='Month', columns='Store', values='Sales')
line_df.plot(kind='line', marker='o')
plt.title('Monthly Sales by Store')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

5. Bonus

Create a bar chart showing total sales per store.

Add data labels on top of each bar.

Save this chart as an image ( sales_chart.png ).

```python

total_sales = df.groupby('Store')['Sales'].sum()

plt.figure(figsize=(6, 4))
bars = plt.bar(total_sales.index, total_sales.values, color='pink')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:,}', ha='center', va='bottom', fontsize=10)
plt.title('Total Sales per Store')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.savefig('sales_chart.png')
plt.show()
```
